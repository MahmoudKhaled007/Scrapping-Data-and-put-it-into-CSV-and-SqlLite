```python
10) Write a sql query to count the number of books that have 'Secret' in thier names and price between 10£ and 25£.
```

```python
11) Write a sql query to get the minimum & maximum price for all the books that have rate 5.
```

```python
mycursor.execute("SELECT MIN(PRICE) AS MINIMUM_PRICE , MAX(PRICE) AS MAXIMUM_PRICE FROM books.book WHERE RATE =5")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
12) Write a sql query to calculate the avg price for all the books that have rate 5.
```

```python
mycursor.execute("SELECT AVG(PRICE) AS AVG_PRICE FROM books.book WHERE RATE = 5")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
13) Write a sql query to sum all book's price that have rate 2 and price between 10 and 40
```

```python
14) Write a sql query to join both books & categories table into one new table containing book_name & category_name & book_rate and book_price.
```

```python
mycursor.execute("SELECT * FROM BOOK  INNER JOIN CATEGORY ON BOOK.CAT_ID = CATEGORY.id")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.
```

```python
mycursor.execute("SELECT COUNT(BOOK_NAME), RATE ,PRICE FROM books.book GROUP BY RATE HAVING COUNT(PRICE) BETWEEN 20 AND 30;")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
16) Write a sql query to calculate how many books each category has having count > 10.
```

```python
mycursor.execute("SELECT COUNT(BOOK_NAME), RATE ,PRICE FROM books.book GROUP BY CAT_ID HAVING COUNT(BOOK_NAME) >10;")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```

```python
17) Write a sql query to get all books with category_name='Music' using subquery.
```

```python
mycursor.execute("SELECT * FROM books.book WHERE CAT_ID=(SELECT ID FROM CATEGORY WHERE NAME ='Music')")

myresult = mycursor.fetchall()

for x in myresult:
    print(x)
```
